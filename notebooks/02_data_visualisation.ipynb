```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les donnes traites
df = pd.read_csv('/Users/omarhaddad/Desktop/ESG_analysis/data/processed/processed_esg_data.csv')

# Visualisation des tendances au fil du temps
plt.figure(figsize=(14, 7))
sns.lineplot(data=df, x='Date', y='Value', hue='Metric', marker='o')
plt.title('Trends in ESG metrics over time')
plt.xticks(rotation=45)
plt.show()

# Boxplot des valeurs ESG par catgorie
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='Category', y='Value')
plt.title('Distribution des valeurs ESG par catgorie')
plt.show()
```
